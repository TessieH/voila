```python
health_fac_full = health_fac_full.replace([np.inf, -np.inf], np.nan).dropna()
gdp = health_fac_full['GDP'].values
adult_mortality = health_fac_full['Adult Mortality'].values
corr,_ = pearsonr(gdp, adult_mortality)
print(f"Pearson correlation coefficient: {corr}")
```

the code above shows a weak negative correlation between GDP and adult mortality overall for all countries through the years 2000-2015.

```python
countries = health_fac_full['Country'].unique()
country_dropdown = widgets.Dropdown(options=countries,value=countries[0],description='Country:')
years = health_fac_full['Year'].unique()
years_dropdown = widgets.Dropdown(options=years,value=years[0],description='Year:')
display(country_dropdown)
display(years_dropdown)

def calculate_pearson_corr(health_fac_full, country, year):
    d = health_fac_full[health_fac_full['Country'] == country][['GDP', 'Adult Mortality']]
    corr, _ = pearsonr(d['GDP'], d['Adult Mortality'])
    print(f"The Pearson correlation between GDP and Adult Mortality for {country} in {year} is: {corr}")
    if corr < -0.8:
        print("There is a strong negative correlation.")
    elif corr < -0.5:
        print("There is a moderate negative correlation.")
    elif corr < -0.2:
        print("There is a weak negative correlation.")
    elif corr < 0.2:
        print("There is no significant correlation.")
    elif corr < 0.5:
        print("There is a weak positive correlation.")
    elif corr < 0.8:
        print("There is a moderate positive correlation.")
    else:
        print("There is a strong positive correlation.")
button = widgets.Button(description='Calculate correlation')

def on_button_clicked(_):
    country = country_dropdown.value
    year = years_dropdown.value
    calculate_pearson_corr(health_fac_full, country, year)
button.on_click(on_button_clicked)
display(button)
```
