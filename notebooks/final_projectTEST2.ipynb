```python
def select_country_data(country_name, df):
  '''Takes iso_code and df as input. Returns the rows in
  df belonging to country with iso_code'''
  
  return RQ_df_nomiss_done.loc[RQ_df_nomiss_done['Country'] == country_name, ]
```

DATA PER COUNTRY

```python
interact(select_country_data, country_name = widgets.Dropdown(options=RQ_df_nomiss_done['Country'].unique()), df=fixed(RQ_df_nomiss));
```

```python
# solution 5.1
import altair as alt

def visualize_country_data(country_name, df):

  # 1. First we use the select_country_data function to select the subset
  # of df (here the covid data) with the specific iso_code
  data_to_viz = select_country_data(country_name, df)

  # 2. use altair to create a chart (save it as variable called chart)
  # the chart should be: a line chart,
  # with 'yearmonth(date):T' on x-axis and
  # 'max(total_vaccinations):Q' on y axis
  # showing location in the tooltip
  ### Your code here to create the chart
  chart = alt.Chart(data_to_viz).mark_line().encode(
      x='Year',
      y='GDP:Q',
      tooltip='Country'
  )

  # display the chart object you created
  display(chart)

  # NB: most commonly a function would 'return' something.
  #  Yet, this gives an issue later on when using interactive_output.
  #  Therefore, we use display(chart) instead of return chart here.
```

```python
interact(visualize_country_data,
         country_name = widgets.Dropdown(options=RQ_df_nomiss_done['Country'].unique()), df=fixed(RQ_df_nomiss));
```

VISUAL

```python


# NEW: adjust visualize_country_data to take y_var as input
def visualize_country_data(country_name, y_var, df):

  data_to_viz = select_country_data(country_name, df)

  chart = alt.Chart(data_to_viz).mark_line().encode(
      x='Year',
      y=f'max({y_var}):Q', # NEW: add y_var to y-axis
      tooltip='Country'
  )

  display(chart)


# define dropdown widget
country_dropdown = widgets.Dropdown(options=RQ_df_nomiss_done['Country'].unique())
# NEW: define a widget for y_var
y_var_dropdown = widgets.Dropdown(options=['GDP', 'Adult Mortality'])

# define the interactive_output element (with function and keyword arguments)
# NEW: add y_var and its widget to the keyword arguments
iout = widgets.interactive_output(visualize_country_data, {'country_name': country_dropdown,
                                                           'y_var': y_var_dropdown,
                                                           'df': fixed(RQ_df_nomiss_done)})

# NEW: add the widget to the layout (I chose to put both widgets together in VBox)
widgets.HBox([iout, widgets.VBox([country_dropdown, y_var_dropdown])])
```
